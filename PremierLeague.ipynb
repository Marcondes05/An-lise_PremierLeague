```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2324/E0.csv?authuser=0")
df
```

```python
# Lista com o nomes dos times
times = df.HomeTeam.unique()
times
```

```python
#Mostrando todas as colunas do df
df.columns.to_list()
```

```python
#Selecionando as colunas que vamos utilizar
colunas = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC']
```

```python
df = df[colunas]
df
```

```python
# Criando as varivéis Dummies na coluna resultado final

df['HomeWin'] = [1 if x == 'H' else 0 for x in df['FTR']]
df['Draw'] = [1 if x == 'D' else 0 for x in df['FTR']]
df['AwayWin'] = [1 if x == 'A' else 0 for x in df['FTR']]
df
```

```python
df.columns.to_list()
```

```python
#reorganizando as colunas
colunas = ['Date', 'HomeTeam', 'AwayTeam', 'HomeWin', 'Draw', 'AwayWin',
           'FTHG', 'FTAG', 'HTHG', 'HTAG', 'HTR',
           'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC']
```

```python
df = df[colunas]
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df
```

Listando todos os times e criando um novo dataframe

```python
premierdf = pd.DataFrame(times, columns=["Equipe"])
premierdf
```

**Exercicio 1**
Criando uma nova coluna de vitorias

```python
premierdf['V'] = 0
```

```python
for time in times:
  premierdf.loc[premierdf.Equipe == time, 'V'] = df.loc[((df['HomeTeam'] == time) & (df['HomeWin'] == 1) | (df['AwayTeam'] == time) & (df['AwayWin'] == 1))].shape[0]
premierdf
```

```python
fig, ax = plt.subplots()
ax.bar(premierdf.Equipe, premierdf.V, color = 'blue')
ax.set_xlabel("Equipes")
ax.set_ylabel('Vitórias')
ax.set_title('Vitórias por Equipe')
plt.xticks(rotation=90)
ax.set_yticks([i for i in range(0, premierdf["V"].max() + 1)])
plt.show()
```

**Exercicio 2**
Fazendo a comparação de pontos ganhos e pontos disputados


```python
premierdf['J']=0 #quantidade de jogos
premierdf['E']=0 #quantidade de empates
premierdf['D']=0 #quantidade de derrotas
premierdf['P']=0 #quantidade de pontos
premierdf['Aprov']=0 #aproveitamento
```

```python
premierdf.columns.to_list()
```

```python
#Reorganizando as colunas para que fique (time, pontos, jogos, vitorias, empates, derrotas, aproveitamento)
colunas = ['Equipe', 'P', 'J', 'V', 'E', 'D', 'Aprov']
```

```python
premierdf = premierdf[colunas]
premierdf
```

```python
for time in times:
  premierdf.loc[premierdf.Equipe == time, 'J'] = df.loc[(df['HomeTeam'] == time) | (df['AwayTeam'] == time)].shape[0] #Quantidade de jogos
  premierdf.loc[premierdf.Equipe == time, 'E'] = df.loc[((df['HomeTeam'] == time) & (df['Draw'] == 1) | (df['AwayTeam'] == time) & (df['Draw'] == 1))].shape[0] #Quantidade deempates
  premierdf.loc[premierdf.Equipe == time, 'D'] = df.loc[((df['HomeTeam'] == time) & (df['AwayWin'] == 1) | (df['AwayTeam'] == time) & (df['HomeWin'] == 1))].shape[0] #Quantidade dederrotas
  premierdf.loc[premierdf.Equipe == time, 'P'] = (premierdf.loc[premierdf.Equipe == time, 'V']*3) + premierdf.loc[premierdf.Equipe == time, 'E']  #Quantidade de pontos
```

```python
#Calculando o Aproveitamento
for time in times:
  premierdf.loc[premierdf.Equipe == time, 'Aprov'] = ((premierdf.loc[premierdf.Equipe == time, 'P']) / (premierdf.loc[premierdf.Equipe == time, 'J']*3)*100)
```

**Exercicio 3**
Completando o restande da tabela

```python
premierdf['GM']=0 #gols marcados
premierdf['GS']=0 #gols sofridos
premierdf['SG']=0 #saldo de gols
premierdf['#']=0 #posição
```

```python
colunas = ['#','Equipe','P', 'J', 'V', 'E', 'D', 'GM', 'GS', 'SG']
```

```python
premierdf = premierdf[colunas]
premierdf
```

```python
for time in times:
  premierdf.loc[premierdf.Equipe == time, 'GM'] = df.loc[(df['HomeTeam'] == time)]['FTHG'].sum() + df.loc[(df['AwayTeam'] == time)]['FTAG'].sum()
  premierdf.loc[premierdf.Equipe == time, 'GS'] = df.loc[(df['HomeTeam'] == time)]['FTAG'].sum() + df.loc[(df['AwayTeam'] == time)]['FTHG'].sum()
  premierdf.loc[premierdf.Equipe == time, 'SG'] = premierdf.loc[premierdf.Equipe == time, 'GM'] - premierdf.loc[premierdf.Equipe == time, 'GS']
```

```python
premierdf.sort_values(by=['P', 'SG', 'V'], ascending=False, inplace=True)
premierdf.reset_index(drop=True, inplace=True)
pos = np.arange(1, times.size+1)
premierdf['#'] = pos
```

**Exercicio 4** - Grafico rodada por rodada


```python
teams = df["HomeTeam"].unique()
colunas_resultados = ["HomeTeam", "AwayTeam",'HomeWin', 'Draw', "AwayWin", "FTHG", "FTAG"]
df2 = df[colunas_resultados]
colunas_evolucao_pontos = ["Time"]

# Agrupa os dados do DataFrame df2 pelos times que jogaram em casa.
jogos_times = df2.groupby("HomeTeam")
semanas = (jogos_times["AwayTeam"].count()+jogos_times["HomeTeam"].count()).min()

for i in range(0, semanas):
  colunas_evolucao_pontos.append(f'Semana{i}')
colunas_evolucao_pontos
tabela_evolucao_pontos = []

for team in teams:
  time_evolucao = [team]
  pontos = 0

  for index, row in df2.iterrows():
    if(len(time_evolucao) < len(colunas_evolucao_pontos)):
      if (row["HomeTeam"] == team):
        if (row["HomeWin"] == 1): pontos += 3
        elif (row["Draw"] == 1): pontos += 1
        time_evolucao.append(pontos)
      if (row["AwayTeam"] == team):
        if (row["AwayWin"] == 1): pontos += 3
        elif (row["Draw"] == 1): pontos += 1
        time_evolucao.append(pontos)

  tabela_evolucao_pontos.append(time_evolucao)

tabela_evolucao_pontos = pd.DataFrame(tabela_evolucao_pontos, columns=colunas_evolucao_pontos)
tabela_evolucao_pontos
```

```python
fig, ax = plt.subplots()

for index, linha in tabela_evolucao_pontos.iterrows():

  # Plotando a evolução dos pontos de cada time ao longo das semanas
  ax.plot(linha.iloc[1:], label=linha["Time"])

ax.tick_params('x',labelrotation=90)
ax.set_xlabel("Semanas")
ax.set_ylabel("Pontos")
ax.set_title("Evolução dos pontos de cada time ao longo das semanas")
ax.grid(color="#edebeb")
fig.legend(bbox_to_anchor=(1.3, 1))
```

Exercicio 5 - posição por rodada

```python
tabela_evolucao_posicao = tabela_evolucao_pontos

for semana in tabela_evolucao_posicao.columns[1:]:

 # Ordenando a tabela pela pontuação da semana atual em ordem decrescente
  tabela_evolucao_posicao = tabela_evolucao_posicao.sort_values(semana, ascending=False)
  tabela_evolucao_posicao = tabela_evolucao_posicao.reset_index(drop=True)
  tabela_evolucao_posicao = tabela_evolucao_posicao.sort_values(semana, ascending=False)

 # Atualizando a coluna da semana atual para refletir a posição do time na tabela
  tabela_evolucao_posicao[semana] = tabela_evolucao_posicao.index+1
tabela_evolucao_posicao
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

for index, linha in tabela_evolucao_posicao.iterrows():
  ax.plot(linha.iloc[1:], label=linha["Time"])
ax.tick_params('x',labelrotation=90)

ax.set_xlabel("Semanas")
ax.set_ylabel("Posição")
ax.set_title("Evolução de posições")

ax.set_yticks([x for x in range(1, len(tabela_evolucao_posicao))])
ax.invert_yaxis()
ax.grid(color="#edebeb")
fig.legend(bbox_to_anchor=(1.05, 1))
```
